# Enter the job title you are looking to search for

```python
print("Hello!")
time.sleep(1)
print("Welcome to our automated web crawling system  ")
time.sleep(1)
search_words =[]
search_word = input("Please ENTER the jobs you're looking for one by one ")
search_words.append(search_word.strip())
while(True):
    search_word = input("ENTER job name if you want to continue 0 otherwise ")
    if(search_word == '0'):
        break
    search_words.append(search_word.strip())
search_words
```

- Instead of going to the page and sending the job title name to get the page, we can pass the job title to the get_link function, which will return the link page.

```python
def get_link(keyword):

    link= "https://remote.co/remote-jobs/search/?search_keywords="
    data = keyword.split()
    if len(data)== 1:
        return link+data[0]
    else:
        word = ""
        for i in data :
            word +=i
            word +='+'

        return link + word[:-1]
```

 - Each page has lots of links to jobs.

```python
Links = []
for search in search_words:
    driver = webdriver.Chrome(path)
    driver.get(get_link(search))
    time.sleep(20)
    popup = driver.find_element_by_css_selector("#om-oqulaezshgjig4mgnmcn-optin > div > button > svg").click()
    while(True):
        try:
            time.sleep(1)
            data = driver.find_element_by_class_name('load_more_jobs')
            data.click()
        except:
            html = driver.page_source
            soup = BeautifulSoup(html)
            break

    Element_Links = soup.find_all(class_ = 'font-weight-bold larger stretched-link',href=True)
    for i in Element_Links:
        Links.append("https://remote.co"+i.attrs["href"])



```

```python
for i in Links:
    print(i)
```

### Creating data frame

```python
df = pd.DataFrame(columns=["JobTitle","CompanyName","Salary","time","Enrollment_status","Location","JobDescription"])
df.head()
```

```python
for link in Links:
    #for each link - get html page scource - and get job imformation
    driver.get(link)
    HTML_job = driver.page_source
    Soup = BeautifulSoup(HTML_job)
    # job inforamtion
    PostedTime = Soup.find(class_ ="date_sm").get_text()
    JobDescription = Soup.find(class_ ="job_description").get_text()
    Location = Soup.find(class_ ="location_sm").get_text()
    try:
        Salary = Soup.find(class_ ="salary_sm").get_text()
    except:
        Salary = ""

    Enrollment_status = Soup.find(class_ ="tags_sm").get_text()
    CompanyName = Soup.find(class_ ="co_name").get_text()
    JobTitle= Soup.find(class_ ="font-weight-bold").get_text()
    df.loc[len(df.index)] = [JobTitle,CompanyName,Salary,PostedTime,Enrollment_status,Location,JobDescription]
    #onejobdetails = [JobTitle,CompanyName,Salary,PostedTime,Enrollment_status,Location,JobDescription]

    time.sleep(1)
```
